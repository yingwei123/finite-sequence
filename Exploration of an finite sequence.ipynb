# Section one
Goal : I wanted to see the numbers that will lead to the sequence's length to increase by one

```python
#Function that gets the sequence from x to y
def getSeq(x,y):
    #empty list to store sequence
    los = []
    #empty list to store number
    lon = []
    #loop from x to y
    for i in range (x,y+1):
        #get the sequence of i
        seq = full_sequence(i)
        #if the sequence is not None then add it to it's respected list
        if(seq !=None):
            los.append(full_sequence(i))
            lon.append(i)
    #return the 2 list
    return los,lon
```

**Define getSeq(x,y)**
getSeq(x,y) is a function that takes in 2 variables, x and y and returns a list of full_sequences of each number from x to y along with its number in los,lon respectively

```python
#Function that Finds the pattern for increasing sequence length by 1
def increasingSequenceLength(x,y):
    #gets the list of sequence and numbers
    los,lon = getSeq(x,y)
    #to store the seq, number, and length of seq
    pattern1 =[]
    #counter to hold the value of the current length to look for
    holder = 0
    for i in range (len(los)):
        #if the length is equal to the holder value add it to pattern1 and find the next seq to the pattern
        if(len(los[i]) == holder):
            arrToAdd = []
            arrToAdd.append(los[i])
            arrToAdd.append(lon[i])
            arrToAdd.append(len(los[i]))
            pattern1.append(arrToAdd)
            holder = holder +1
    return pattern1
```

**Define increasingSequenceLength(x,y)**
\
This function returns a list of a list of sequences as well as its length between x and y

Goal: Print out the first index of all even numbers of the sequence to find a common pattern


```python
#function to get all the first index of even number sequence
def getAllEvenSeq(x,y):
    #gets the sequence of x to y
    los,lon = getSeq(x,y)
    #list for even number sequences
    listOfEvenSeq=[]
    #list of even numbers
    lofEven = []
    #length of the sequence
    length = []
    for i in range (len(los)):
        #mod it by 2 to see if it is even, if it is add it to its respective array
        if(lon[i]%2==0):
            ##add the first index of the sequence 
            listOfEvenSeq.append(los[i][0])
            lofEven.append(lon[i])
            length.append(len(los[i]))
    return listOfEvenSeq, lofEven, length
```

# Section 5
Goal : Find the length of sequences for prime and non prime numbers

```python
#determines if a number is prime
def isPrime(x,y):
    for i in range (2,y):
        #x is divisible by any of 2-9 then it is not a prime number
        if(x %i ==0 and x != i):
            return False
    return True
```

**Define isPrime(x,y)**
isPrime(x,y) returns true if x is a prime number, it will check if x is divisible by any number from 2 to y as long as 
it is divisible by more than 1 number it will return false otherwise it will return true

**Explore 5**
\
Explore the length of sequence of numbers that are prime and non prime

```python
#seperate and graph prime and non-prime numbers
los,lon = getSeq(0,1000)
#Array used to store prime numbers
prime =[]
#Array used to store the prime number sequence lengths
primelen =[]
#Array used to store non prime numbers
nonprime = []
#Array used to store non prime number sequences
nonprimelength = []
for a in range (len(los)):
    #if the number is prime then add it to the prime array
    if(isPrime(lon[a],1000)):
        prime.append(lon[a])
        primelen.append(len(los[a]))
    else:
        #else Add the number to the non prime array
        nonprime.append(lon[a])
        nonprimelength.append(len(los[a]))
        
plt.title("Length of sequence of prime numbers")
plt.plot(prime, primelen)
plt.xlabel("Prime Numbers")
plt.ylabel("Length")
plt.show() 

plt.title("Length of Non Prime numbers")
plt.plot(nonprime, nonprimelength)
plt.xlabel("Non Prime Numbers")
plt.ylabel("Length")
plt.show() 
```
